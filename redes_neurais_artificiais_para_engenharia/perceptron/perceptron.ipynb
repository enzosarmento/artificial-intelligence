2. Mostre por intermédio de gráficos ilustrativos como pode ocorrer a
instabilidade no processo de convergência do Perceptron quando da utilização
de valores inapropriados para a taxa de aprendizado.

````python
#R: Utilizando o matplotlib para plotar gráficos que ilustram a instabilidade do Perceptron com uma taxa de aprendizado alta e a estabilidade com uma taxa de aprendizado adequada.

import matplotlib.pyplot as plt
import numpy as np

# Create synthetic data for two classes
np.random.seed(42)
# Class A data
class_a_x1 = np.random.normal(2, 1, 30)
class_a_x2 = np.random.normal(2, 1, 30)

# Class B data
class_b_x1 = np.random.normal(6, 1, 30)
class_b_x2 = np.random.normal(6, 1, 30)

# Create the plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# --- Plot 1: High Learning Rate (Unstable) ---
ax1.scatter(class_a_x1, class_a_x2, marker='o', label='Classe A')
ax1.scatter(class_b_x1, class_b_x2, marker='x', label='Classe B')
x_vals = np.linspace(0, 8, 10)

# Simulate oscillating lines
# Epoch 1
w1, w2, theta = 0.5, -0.2, 1.5
ax1.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'r--', label='Época 1')
# Epoch 2
w1, w2, theta = -0.1, 0.4, 1.8
ax1.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'g-.', label='Época 2')
# Epoch 3
w1, w2, theta = 0.6, -0.1, 2.0
ax1.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'm:', label='Época 3')


ax1.set_title('Taxa de Aprendizado Alta (Instabilidade)')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_xlim(0, 8)
ax1.set_ylim(0, 8)
ax1.legend()
ax1.grid(True)

# --- Plot 2: Appropriate Learning Rate (Stable) ---
ax2.scatter(class_a_x1, class_a_x2, marker='o', label='Classe A')
ax2.scatter(class_b_x1, class_b_x2, marker='x', label='Classe B')

# Simulate converging lines
# Epoch 1
w1, w2, theta = 0.1, 0.5, 0.5
ax2.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'r--', label='Época 1')
# Epoch 10
w1, w2, theta = 0.3, 0.3, 2.0
ax2.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'g-.', label='Época 10')
# Epoch 20 (Converged)
w1, w2, theta = 0.5, 0.5, 4.0
ax2.plot(x_vals, (-w1/w2) * x_vals + (theta/w2), 'm:', label='Época 20 (Final)')

ax2.set_title('Taxa de Aprendizado Adequada (Estabilidade)')
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')
ax2.set_xlim(0, 8)
ax2.set_ylim(0, 8)
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig('learning_rate_comparison.png')
````
